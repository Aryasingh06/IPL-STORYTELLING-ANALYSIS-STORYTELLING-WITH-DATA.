###  Dataset Overview
-- The dataset contains match-level and ball-by-ball data of the Indian Premier League.The matches dataset provides 
   overall match information, while the deliveries dataset contains detailed ball-by-ball records.
   
-- The matches dataset contains 636 matches with 18 columns.The deliveries dataset contains ball-by-ball data 
    for all matches.
    
-- Data Consistency Check: The number of unique match IDs in the deliveries dataset is equal to the total number of matches     (636),confirming that ball-by-ball data is available for every match.

```python
#DATA CLEANING
#check duplicate team names
matches['team1'].unique()
```

```python
#handle missing values
matches['umpire1']=matches['umpire1'].fillna('Unknown')
matches.isnull().sum()
```

```python
matches['umpire2']=matches['umpire2'].fillna('Unknown')
matches.isnull().sum()
```

```python
matches['umpire1'].fillna('Unknown', inplace=True)
matches.isnull().sum()
```

```python
#REMOVE DUPLICATES
matches.drop_duplicates(inplace=True)
```

```python
#Check duplicates
matches.duplicated().sum()
```

## Data Cleaning
 Missing values in the dataset were handled appropriately.
 The umpire1 and umpire2 columns were filled with "Unknown", while missing valuesin the winner column were retained  as
 they represent valid match outcomes (tie/no result).Duplicate records were checked and removed to ensure data quality 
 and consistency.

### ANALYSIS + VISUALIZATIONS

```python
#WHICH TEAM WON MOST MATCHES
matches['winner'].value_counts().head(5).plot(kind='bar')
plt.title("Top 5 Teams with Most Wins")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.show()
```

### Mumbai Indians has the highest number of wins, indicating consistent performance across seasons.


```python
# PLAYER OF THE MATCH ANALYSIS
matches['player_of_match'].value_counts().head(5).plot(kind='bar')
plt.title("Top Players of the Match Awards")
plt.xlabel("Players")
plt.ylabel("Number of Awards")
plt.show()
```

### Players like Chris Gayle and YK Pathan  frequently appear in top positions, showing match-winning performances.

```python
#TOSS IMPACT ANALYSIS
matches['toss_winner'].value_counts().plot(kind='bar')
plt.title("Does Toss Winner Also Win Match?")
plt.show()
```

### Winning the toss does not always guarantee winning the match, indicating that match performance            matters more than toss advantage.

```python
#BATTING BOWLING ANALYSIS
#TOP RUN SCORER

top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(5)
top_batsmen.plot(kind='bar')
plt.title("Top 5 Run Scorers")
plt.xlabel("Batsmen")
plt.ylabel("Total Runs")
plt.show()
```

### Sk Raina ranks among the highest run scorers, reflecting his reliability and long-term contribution as a key batsman in the Indian Premier League.

```python
# highest wicket taker
# remove extra runs like run out etc.
wickets = deliveries[deliveries['dismissal_kind'] != 'run out']

top_bowlers = wickets['bowler'].value_counts().head(5)
top_bowlers.plot(kind='bar')
plt.title("Top 5 Wicket Takers")
plt.xlabel("Bowlers")
plt.ylabel("Number of Wickets")
plt.show()
```

### Harbhajan Singh is among the leading wicket-takers in the dataset, highlighting his effectiveness and consistency as a bowler in the Indian Premier League.
